## Analyzing

For the code1 method, the complexity of algorithms is O(N^2). In relation to the code2 method, however, it is O(N). Look at the graphs to compare them.

```python
import numpy as np
import random
import time
import matplotlib.pyplot as plt
import random
```

```python
class Codes():
  def __init__(self, array, targetSum):
    self.array = array
    self.targetSum = targetSum

  # Code 1 -> O(N^2)
  def code1(self):
    for i in range(len(self.array) - 1):              
      firstNum = self.array[i]                        
      for j in range (i+1, len(self.array)):          
        secondNum = self.array[j]                     
        if firstNum + secondNum == self.targetSum:    
          return [firstNum, secondNum]                
    return []                                         

  # Code 2 -> O(N)
  def code2(self):
    sol = {}                                          
    for x in self.array:                              
      y = self.targetSum - x                          
      if y not in sol:                               
        sol[x] = True                                 
      else:                                                                            
        return [y, x]                                 
    return []                                         
```

### Complexy Algorithm

#### Time

It's possible to notice that until a certain size, there's not make much difference between the performance of the two codes. However, as the size of the array increases, this difference becomes visible.

```python
def plot_times(times, g):
  fig, ax = plt.subplots(1,1,figsize=(10,6))
  ax.plot(times,label="Real execution time")
  ax.set_xlabel("Array Size")
  ax.set_ylabel("Cost")

  size = len(times)

  if g == 1:
    z = np.polyfit(range(0, size),times, 1)
  elif g == 2:
    z = np.polyfit(range(0, size),times, 2)
  p = np.poly1d(z)

  ax.plot(range(0, size),p(range(0, size)),label="Time complexity model")
  ax.legend()
  plt.show()

def gen_input(length):
    return [random.randint(-1000, 1000) for _ in range(length)]

def time_alg(code, g):
  times =[]
  for length in range(1, 5000):
      targetSum = length
      values = gen_input(length)
      codes = Codes(values, targetSum)
      if code == 1:
        start = time.time()
        codes.code1()
        end = time.time()
      elif code == 2:
        start = time.time()
        codes.code2()
        end = time.time()
      times.append(end - start)
  if g == 1:
    plot_times(times,1)
  elif g == 2:
    plot_times(times,2)
```

### Code 1 - O(N^2)

```python
time_alg(1,2)
```

### Code 2 - O(N)

```python
time_alg(2,1)
```
